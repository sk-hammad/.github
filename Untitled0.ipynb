```python
import matplotlib.pyplot as plt
import numpy as np

# Load the results file
results_file = "/content/yolov5/runs/train/exp2/results.csv"  # Update path if needed
data = np.loadtxt(results_file, delimiter=',', skiprows=1)

# Columns: [epoch, train_loss, val_loss, P, R, mAP_0.5, mAP_0.5:0.95]
epochs = data[:, 0]
train_loss = data[:, 1]
val_loss = data[:, 2]
mAP50 = data[:, 5]
mAP5095 = data[:, 6]

# Plotting
plt.figure(figsize=(12, 6))

# Loss plot
plt.subplot(1, 2, 1)
plt.plot(epochs, train_loss, label="Train Loss")
plt.plot(epochs, val_loss, label="Validation Loss", linestyle="--")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Curve")
plt.legend()

# mAP plot
plt.subplot(1, 2, 2)
plt.plot(epochs, mAP50, label="mAP@0.5")
plt.plot(epochs, mAP5095, label="mAP@0.5:0.95", linestyle="--")
plt.xlabel("Epoch")
plt.ylabel("mAP")
plt.title("mAP Curve")
plt.legend()

plt.tight_layout()
plt.show()
```
